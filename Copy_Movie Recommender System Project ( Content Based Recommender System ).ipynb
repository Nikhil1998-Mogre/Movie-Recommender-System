- we want to convert tag into vector
- For this we are using Bag of Word approach
- where we are combining all tag and finding most frequent words are used( 5000 words)
- Then we have matrix for each movies with 5000 columns for each movies and how many times that words repeated in that tag
- Then make vector on that basis
- then cal. cosine similarity to find closest movies to that.

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

### <font color='Red'> English Stopword </font>

```python
stopwords.words('english')            ### stopword in english
```

### <font color='Red'> Getting 5000 common words </font>

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new['tag']).toarray()            ## we will get sparse matrix we convert to array use toarray
```

```python
cv.fit_transform(new['tag']).toarray().shape         ### For each movies we have number of times the most frquent words repeat in that movie tag
```

```python
vectors        ### we converted movies tag into vectors
```

```python
cv.get_feature_names()[47]                ### this(5000 words ) are the most frequent words used in large tag( when we combine all tags)
```

### <font color='Red'> Remove Similar words from tag </font>

```python
### But there is problem like of same word repeat like action and actions, actor and actors

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
ps.stem('actors')
```

```python
def stem(v):
    y=[]
    for i in v.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tag'] = new['tag'].apply(stem)
```

```python
new['tag'][0]
```

```python
vectors = cv.fit_transform(new['tag']).toarray()   ## after doing stem to remove similar words from 5000 words
```

```python
cv.get_feature_names()                      ### now actor are there no actors
```

## cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_dis = cosine_similarity(vectors)
```

```python
sim_dis[0]                   ## so similarity with himself is 1 and with other 4799 movies is given
```

```python
sim_dis.shape
```

```python
sorted(list(enumerate(sim_dis[0])),reverse=True,key = lambda x : x[1])     ### sorting on basis of angle not on index
```

# <font color='Orange'> Movies Recommendations </font>

```python
def recom(movie):
    movie_index = new[new['title']==movie].index[0]
    distance = sim_dis[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x : x[1])[1:20]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)            ## give movie title
       
```
